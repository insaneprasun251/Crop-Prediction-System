```python
import pandas as pd
import numpy as np
```

```python
crops_df = pd.read_csv("Crop_recommendation.csv")
```

```python
crops_df.info()
```

```python
crops_df.describe()
```

```python
crops_df.dropna(subset=['label'], inplace=True)
```

```python
crops_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = crops_df.iloc[:, :-1]
Y = crops_df.iloc[:, -1] 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(X)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
base_model = RandomForestClassifier(random_state=42, n_jobs=-1).fit(X_train, Y_train)
```

```python
base_train_acc = base_model.score(X_train, Y_train)
base_test_acc = base_model.score(X_test, Y_test)
```

```python
base_accs = base_train_acc, base_test_acc
base_accs
```
